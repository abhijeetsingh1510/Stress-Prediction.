```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import numpy as np

# Load the dataset
data = pd.read_csv('/content/Copy--edited_1_.csv')

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Scale features
scaled_features = scaler.fit_transform(data[['TEMPERATURE(K)', 'STRAIN RATE', 'true strain']])
data[['TEMPERATURE(K)', 'STRAIN RATE', 'true strain']] = scaled_features

# Define a PyTorch dataset
class ForgingDataset(Dataset):
    def __init__(self, data):
        self.X = torch.tensor(data[['TEMPERATURE(K)', 'STRAIN RATE', 'true strain']].values, dtype=torch.float32)
        self.y = torch.tensor(data['true stress'].values, dtype=torch.float32).view(-1, 1)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# Create dataset and dataloader
dataset = ForgingDataset(data)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# Define the neural network with an additional hidden layer
class SimpleNN(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2, output_size):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size1)
        self.fc2 = nn.Linear(hidden_size1, hidden_size2)
        self.fc3 = nn.Linear(hidden_size2, output_size)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Initialize network, criterion, and optimizer
input_size = 3
hidden_size1 = 200
hidden_size2 = 200  # Additional hidden layer size
output_size = 1
model = SimpleNN(input_size, hidden_size1, hidden_size2, output_size).to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train the network
num_epochs = 100
for epoch in range(num_epochs):
    for batch_idx, (data, targets) in enumerate(dataloader):
        data = data.to(device)
        targets = targets.to(device)

        # Forward pass
        outputs = model(data)
        loss = criterion(outputs, targets)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

print('Finished Training')

# Evaluate the model
model.eval()  # Set the model to evaluation mode

# Initialize lists to store predictions and targets
predictions = []
targets_list = []

# Iterate through the dataloader for evaluation
for data, targets in dataloader:
    data = data.to(device)
    targets = targets.to(device)

    # Forward pass
    with torch.no_grad():
        outputs = model(data)

    # Convert predictions and targets back to numpy arrays
    predictions.extend(outputs.cpu().numpy())
    targets_list.extend(targets.cpu().numpy())

# Convert lists to numpy arrays
predictions = np.array(predictions)
targets_list = np.array(targets_list)

# Calculate Mean Squared Error (MSE)
mse = np.mean((predictions - targets_list) ** 2)
print(f'Mean Squared Error (MSE): {mse:.4f}')

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')

# Calculate R-squared (R2 score)
ss_res = np.sum((targets_list - predictions) ** 2)
ss_tot = np.sum((targets_list - np.mean(targets_list)) ** 2)
r2_score = 1 - (ss_res / ss_tot)
print(f'R-squared (R2 score): {r2_score:.4f}')

```
